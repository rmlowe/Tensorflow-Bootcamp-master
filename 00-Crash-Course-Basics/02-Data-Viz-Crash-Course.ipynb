```python
x = np.arange(0,10)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title("TITLE")
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
```

```python
plt.plot(x,y,'r--')
plt.xlim(2,4)
plt.ylim(10,20)
plt.title("Zoomed")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```
